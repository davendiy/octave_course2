```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline


def task_g(y, x):
    """ Function g(y, x) = y^2 - y in y' = g(y, x)
    
    The x and y swapped in order to using this function as a parameter of 'odeint'
    """
    return y**2 - y


def task_f(x, y):
    """ Function f(x, y) = y^2 - 2y + x in y' = f(x, y)
    """
    return y**2 - 2*y + x


def task_f2(y, x):
    """ Function f(x, y) = y^2 - 2y + x in y' = f(x, y)
    
    The x and y swapped in order to using this function as a parameter of 'odeint'
    """
    return y**2 - 2*y + x

```

## Task 2 (variant 7)

Let there is differential equation $\frac{dy}{dx} = y^2 - y$

### 1) plot the vector field at the points  $0, \pm1, \pm2, \pm3, \pm4$ 

```python
y = np.arange(-4, 5, 1)
vectors = task_g(y, 0)

# ----------------------------------- plotting vectors under each other -----------------------
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
x = np.arange(-10, 25)
axs[1].set_title('The vectors are arranged under\n each other in order to except overlays')
axs[1].plot(x, np.zeros_like(x))
axs[1].quiver(y, y, vectors - y, np.zeros_like(vectors), angles='xy', scale_units='xy', scale=1)

# ----------------------------------- plotting vectors on the one line -----------------------
fig = plt.figure(figsize=(10, 5))
x = np.arange(-10, 25)
axs[0].set_title("The vectors are arranged as they should be.")
axs[0].plot(x, np.zeros_like(x))
axs[0].quiver(y, np.zeros_like(y), vectors - y, np.zeros_like(vectors), angles='xy', scale_units='xy', scale=1)

plt.show()
```

### 2) plot the set of functions that satisfy the equation

```python
fig = plt.figure(figsize=(12, 12))      # set the size of plot

# ------------------------------ plotting the directions field -----------------------------
x = np.linspace(-5, 10, 32)
y = np.linspace(-2, 2, 32)

U = [1 for _ in range(32)]
V = [task_g(i, 0) for i in y]

XX, YY = np.meshgrid(x, y)
UU, VV = np.meshgrid(U, V)

plt.quiver(XX, YY, UU, VV, units='xy', width=0.009, scale=10)

# ----------------------------- plotting the set of solutions ------------------------------
x = np.arange(-5, 10, 0.1)
for y0 in [-2, -1, -1/2, 1/2, 3/4, 5/6, 0, 0.9, 0.99, 1]:
    plt.plot(x, odeint(task_g, y0, x))

plt.show()
```

### 3) find the points of the inflection of functions

We can find points of the infletion of functions that satisfy the equation $y' = f(x, y)$ solving the equation
$f'_{x} + f'_{y}f = 0$

After subtituting the function from 7 variant we have:
$$0 + (2y - 1)(y^2 - y) = 0$$
$$y(2y-1)(y-1) = 0$$
$$y = 0, y = \frac{1}{2}, y = 1$$

```python

fig = plt.figure(figsize=(12, 12))

# ------------------------------ plotting the directions field -----------------------------
x = np.linspace(-5, 10, 32)
y = np.linspace(-2, 2, 32)

U = [1 for _ in range(32)]
V = [task_g(i, 0) for i in y]

XX, YY = np.meshgrid(x, y)
UU, VV = np.meshgrid(U, V)

plt.quiver(XX, YY, UU, VV, units='xy', width=0.009, scale=10)

# ----------------------------- plotting the set of solutions ------------------------------
x = np.arange(-5, 10, 0.1)
for y0 in [-2, -1, -1/2, 1/2, 3/4, 5/6, 0, 0.9, 0.99, 1]:
    plt.plot(x, odeint(task_g, y0, x), color='b')

plt.plot(x, (np.exp(-9.8 + x) + 1), color='b', label='solutions')   # the last solution with the label

# ---------------------- plotting the lines where inflections situated are -----------------
plt.plot(x, np.zeros_like(x) + 1/2, color='black', label='infections')
plt.plot(x, np.zeros_like(x), x, np.zeros_like(x) + 1, color='black')
plt.legend(loc=2)
plt.show()
```

### 2) describe in analytical way and plot all the maximums of functions that satisfy the equation

To describe all the maximums we need to solve the equation with condition
$$y' = f(x, y) = 0,\quad  y'' = f'_x + f'_y f \le 0$$
$$y^2 - 2y + x = 0, \quad   2y - 2 + (y^2 - 2y + x) \le 0$$
$$y = 1\pm \sqrt{1 - x}, \quad  y^2\le 2 - x, \quad (1)$$


```python
array = np.arange(-5, 5, 0.25)

# ------------------------------ plotting the directions field -----------------------------
XX, YY = np.meshgrid(array, array)

UU = np.cos(np.arctan(task_f(XX, YY)))   # set the x and y coordinates for each vector as a cos and sin
VV = np.sin(np.arctan(task_f(XX, YY)))   # respectively in order to get vectors of unit length
fig = plt.figure(figsize=(15, 15))    

plt.ylim(-5, 5)   # set the border of cartesian plane
plt.quiver(XX, YY, UU, VV, width=0.001)

# -------------------------- plotting solutions between special points -----------------------------
arr1 = np.arange(-5, -1.2, 0.2)  

# upper parabola (there are infinite number of different solutions that coincide with this 
# parabola, so we can't find the one any solution by numerical method from these points)
arr2 = 0.85 + np.sqrt(1-arr1)   # thus the initial states must be a little bit lower than parabola
for (x0, y0) in zip(arr1, arr2):    
    x = np.arange(x0, 3.1, 0.1)    # parameters have been chosen by trial and error method
    plt.plot(x, odeint(task_f2, y0, x), color='b')

# -------------------------- plotting solutions above special points -----------------------------
arr1 = np.arange(-5, 0, 0.2)

# in the same way the initial states must be a little bit greater than parabola
arr2 = 1.1 + np.sqrt(1-arr1) 
for (x0, y0) in zip(arr1, arr2):
#     print(x0)
    if x0 < -3:
        x = np.arange(x0, x0 + 0.8, 0.1)   # also by the trial and error method
    else:
        x = np.arange(x0, x0 + 1, 0.1)
    plt.plot(x, odeint(task_f2, y0, x), color='b')


# ------------------------- plotting solutions below the special points --------------------------------
arr1 = np.arange(-5, 0.6, 0.2)
# print(arr1)
for x0 in arr1:
    x = np.arange(x0, x0 + 3, 0.1)
    plt.plot(x, odeint(task_f2, -5, x), color='b')

# ------------------------------ plotting the curves of extremums --------------------------------------
x = np.arange(-5, 2, 0.1)
plt.plot(x, 1 + np.sqrt(1 - x), color='black', label='extremums')
plt.plot(x, 1 - np.sqrt(1 - x), color='black')

# ------------------------------ plotting the bounds from (1) ------------------------------------------
x = np.arange(-5, 2.5, 0.1)
plt.plot(x, np.sqrt(2 - x), color='red', label='bounds for y (extremums inside only)')
plt.plot(x, -np.sqrt(2 - x), color='red')
plt.legend(loc=4)
plt.show()
```

### 3)  and 4) plot the isoclines with coefficient of an angle: $k = 1, 2, 3$ and direction field at random points of each isocline

Isoclines are described by equation $f(x, y) = k$. Let we substitute function from variant 7 and express any variable through another:

$$k = y^2 - 2y + x \Rightarrow x = -y^2 + 2y + k$$

```python
array = np.arange(-5, 5.5, 0.25)

fig = plt.figure(figsize=(12, 12))
plt.xlim(-14, 5)

# ------------------------------ plotting the directions field -----------------------------
XX, YY = np.meshgrid(np.arange(-14, 5, 0.25), array)

UU = np.cos(np.arctan(task_f(XX, YY)))   # set the x and y coordinates for each vector as a cos and sin
VV = np.sin(np.arctan(task_f(XX, YY)))   # respectively in order to get vectors of unit length

plt.quiver(XX, YY, UU, VV, width=0.001)

# -------------------------------- plotting the isoclines ----------------------------------
k = [1, 2, 3]   # coefficients

y = np.arange(-3, 5, 0.2)   
for tmp_k in k:
    x = -y**2 + 2*y + tmp_k
    plt.plot(x, y)
    
# --------------------------- plotting the directions field at isoclines ----------------------\\
y = np.arange(-3, 5)       # 
for tmp_k in k:
    x = -y**2 + 2*y + tmp_k
    U = 1
    V = task_f(x, y)
    plt.quiver(x, y, U, V, width=0.005)
plt.show()
```

### 5) show the regions where solutions grow up and down 

$y$ grows up $\Leftrightarrow  y' \ge 0 \Rightarrow f(x, y) \ge 0$

$$y^2 - 2y + x >= 0$$
$$y \ge 1 + \sqrt{1 - x},\, or\, y \le 1 - \sqrt{1 - x}, \, x \in [-\infty, 1] \, (*)$$

Thus all the solutions will grow up $\Leftrightarrow$ $y$ and $x$ satisfy the inequality $(*)$ or $x > 1$ 

```python
array = np.arange(-5, 5, 0.25)

# ------------------------------ plotting the directions field -----------------------------
XX, YY = np.meshgrid(array, array)
UU = np.cos(np.arctan(task_f(XX, YY)))
VV = np.sin(np.arctan(task_f(XX, YY)))
fig = plt.figure(figsize=(15, 15))

plt.quiver(XX, YY, UU, VV, width=0.001)

# ---------------------- plotting the curve, that splitting plane at 2 subplanes -----------
array = np.arange(-5, 2, 0.1)
plt.plot(array, 1 - np.sqrt(1 - array), color='red', 
         label='solutions are grow up otside of this bound')
plt.plot(array, 1 + np.sqrt(1 - array), color='red')

plt.legend(loc='best')
plt.show()
```

### 8) realise the Adams method for solving the differential equations and plot the approximate solution that goes through the (0, 0)



```python
from scipy.integrate import RK45


def adams_meth(func, xn, y0):
    """ Adams method for solving the differential equations
    
    :param func: callable, vectorized, func(y, x) is f(y, x) from y' = f(y, x)
    :param xn: array-like, values of x-s
    :param y0: y(xn[0]) - value of y at xn[0]
    :return: np.array - values of y at xn
    """
    if len(xn) <= 4:
        return odeint(func, xn, y0)    # if length <= 4 is just Runge-Kutta
    else:
        yn = np.zeros_like(xn)
        yn[:4] = odeint(func, y0, xn[:4])[0]    # the first 4 elements calculated by RK
        for i in range(3, len(xn)-1):
            h = xn[i+1] - xn[i]            # step
            
            # approximate forecast
            tmp_y = yn[i] + h / 24 * (- 9 * func(yn[i-3], xn[i-3]) 
                                      + 37 * func(yn[i-2], xn[i-2]) 
                                      - 59 * func(yn[i-1], xn[i-1]) 
                                      + 55 * func(yn[i], xn[i],))
            
            # exact forecast
            yn[i+1] = yn[i] + h / 24 * (func(yn[i-2], xn[i-2]) 
                                        - 5 * func(yn[i-1], xn[i-1])
                                        + 19 * func(yn[i], xn[i])
                                        + 9 * func(tmp_y, xn[i+1]))  
        return yn

# ------------------------------ plotting the directions field -----------------------------
array = np.arange(-5, 5, 0.25)

XX, YY = np.meshgrid(array, array)
UU = np.cos(np.arctan(task_f(XX, YY)))
VV = np.sin(np.arctan(task_f(XX, YY)))
fig = plt.figure(figsize=(15, 15))

plt.ylim(-5, 5)
plt.quiver(XX, YY, UU, VV, width=0.001)

# ----------------------- plotting the approximate solution using adams mathod -------------
xn = np.arange(0, 3.4, 0.1)
yn = adams_meth(task_f2, xn, 0)
plt.plot(xn, yn, color='red', label='adams_method')

# -------------------------- plotting the solution using odeint ----------------------------
xn = np.arange(0, 3.4, 0.1)
yn = odeint(task_f2, 0, xn)
plt.plot(xn, yn, '--b', label='built-in function')

plt.legend(loc=2)
plt.show()
```
